```python
#we can see tht r2 score varies largely for train nad testing over model
#we can now try to adjust the hyper parameter
#alpha and gemma hyperparas alpha for regularizing and gemma for kenle function
#0.02,0.18
```

```python
import numpy as np
alpha_values=[1,0.2,0.02,10,8]
gemma_values=[0.01,1.1,0.18,7]
r2=-np.inf
for i in alpha_values:
    for j in gemma_values:
        k_ridge.set_params(alpha=i,gamma=j)
        k_ridge.fit(x_train,y_train)
        y_predict_k=k_ridge.predict(x_test)
        r=r2_score(y_test,y_predict_k)
        if r>r2:
            r2=r
            m=i
            n=j
```

```python
print('best r2 score',r2)
```

```python
k_ridge.set_params(alpha=m,gamma=n)
k_ridge.fit(x_train,y_train)
y_train_predict=k_ridge.predict(x_train)
y_test_predict=k_ridge.predict(x_test)

print('test r2 err when alpha=',m,'gemma=',n)
print('r2 err',r2_score(y_test,y_test_predict))
print('train r2 err when alpha=',m,'gemma=',n)
print('r2 err',r2_score(y_train,y_train_predict))
print(mean_squared_error(y_test,y_test_predict))
mean_squared_error(y_train,y_train_predict)
```

```python
#polynomial regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline 
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
np.random.seed(0)
nus_x=45
x=np.sort(np.random.rand(45))
y=np.sin(2*np.pi*x)+np.random.randn(nus_x)*0.2
```

```python
x
```

```python
y
```

```python
degrees=list([2,7,15])
i=1
s=[]
r2_v=-np.inf
pt.figure(figsize=(20,8))
for d in degrees:
    linear=LinearRegression()
    poly=PolynomialFeatures(d)
    pipe=Pipeline([('features',poly),('linear',linear)])
    x=x.reshape(-1,1)
    pipe.fit(x,y)
    input_test=np.linspace(0,1,50)
    input_test=input_test.reshape(-1,1)
    out_test=np.sin(2*np.pi*input_test)
    #input_test out_test is the actual value
    out_test_pred=pipe.predict(input_test)
    pt.subplot(1,len(degrees),i)
    pt.scatter(x,y,label='training',color='red') #training points
    pt.scatter(input_test,out_test,label='test',color='green')
    pt.plot(input_test,out_test_pred,color='blue',label='pred_out')
    pt.xlabel('features')
    pt.ylabel('result')
    pt.legend()
    tit='degree '+str(d)
    pt.title(tit)
    pt.xticks([])
    pt.yticks([])
    i=i+1
    r=r2_score(out_test,out_test_pred)
    s.append(r)
    if(r>r2_v):
        r2_v=r
        deg=d
```
