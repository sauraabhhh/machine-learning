```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
features=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
features
```

```python
target=data.target
target
```

```python
target=pd.DataFrame(target,columns=['type'])
data=pd.concat([features,target],axis='columns')
```

```python
data.describe()
```

```python
import matplotlib.pyplot as pt
import seaborn as sb
sb.boxplot(features,orient='h')
pt.show()
```

```python
from sklearn.preprocessing import StandardScaler
stan=StandardScaler()
features_scaled=stan.fit_transform(features)
```

```python
features_scaled=pd.DataFrame(features_scaled,columns=features.columns)
```

```python
features_scaled
```

```python
sb.boxplot(features_scaled,orient='h')
pt.show()
```

```python
features_scaled.describe()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features_scaled,target,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_train.describe()
```

```python
import matplotlib.pyplot as pt
import seaborn as sb
sb.boxplot(x_train,orient='h')
pt.show()
```

```python
sb.histplot(x_train['mean radius'],kde=True)
```

```python
med=x_train['mean radius'].median()
import numpy as np
q1=np.quantile(x_train['mean radius'],0.25)
q3=np.quantile(x_train['mean radius'],0.75)
iqr=q3-q1
low=q1-(1.5*iqr)
upp=q3+(1.5*iqr)
outliers=x_train[(x_train['mean radius']<low) | (x_train['mean radius']>upp)]
outliers_index=x_train[(x_train['mean radius']<low) | (x_train['mean radius']>upp)].index
```

```python
outliers_index
```

```python
outliers
```

```python
med=x_train['mean texture'].median()
import numpy as np
q1=np.quantile(x_train['mean texture'],0.25)
q3=np.quantile(x_train['mean texture'],0.75)
iqr=q3-q1
low=q1-(1.5*iqr)
upp=q3+(1.5*iqr)
outliers=x_train[(x_train['mean texture']<low) | (x_train['mean texture']>upp)]
outliers_index=x_train[(x_train['mean texture']<low) | (x_train['mean texture']>upp)].index
```

```python
outliers_index
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(solver='liblinear',penalty='l1')
model.fit(x_train,y_train)
train_predict=model.predict(x_train)
from sklearn.metrics import confusion_matrix,accuracy_score
matrix=confusion_matrix(y_train,train_predict)
accuracy=accuracy_score(y_train,train_predict)
```

```python
print(matrix)
print(accuracy)
```

```python
test_predict=model.predict(x_test)
matrix=confusion_matrix(y_test,test_predict)
accuracy=accuracy_score(y_test,test_predict)
```

```python
print(matrix)
print(accuracy)
```
