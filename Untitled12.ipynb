```python
import pandas as pd 
import matplotlib.pyplot as pt
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
features=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
features
```

```python
target=pd.DataFrame(data.target,columns=['target'])
target
```

```python
dataset=pd.concat([features,target],axis='columns')
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
#perform standardization
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
features_scaled=std.fit_transform(features)
features_scaled=pd.DataFrame(features_scaled,columns=features.columns)
features_scaled
```

```python
i=1
for x in features_scaled:
    pt.subplot(6,5,i)
    pt.hist(features_scaled[x])
    i=i+1
pt.show()    
```

```python
import seaborn as sb
sb.boxplot(features_scaled,orient='h')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features_scaled,target,test_size=0.2,random_state=35)
x_train
```

```python
x_test
```

```python
traing_set=pd.concat([x_train,y_train],axis='columns')
traing_set
```

```python
traing_set.corr()
```

```python
sb.heatmap(traing_set.corr())
```

```python
benign=traing_set[traing_set['target']==0]
```

```python
benign
```

```python
melignant=traing_set[traing_set['target']==1]
melignant
```

```python
pt.scatter(melignant['mean radius'],melignant['mean perimeter'])
pt.scatter(benign['mean radius'],benign['mean perimeter'])
pt.show()
```

```python
from sklearn.svm import SVC
model=SVC(kernel="linear", random_state=35)
model.fit(x_train,y_train)
train_out=model.predict(x_train)
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score,ConfusionMatrixDisplay
test_out=model.predict(x_test)
print('train acc=', accuracy_score(y_train,train_out))
print('train mat=',confusion_matrix(y_train,train_out))
print('f1=',f1_score(y_train,train_out))
print('--------')
print('test acc=', accuracy_score(y_test,test_out))
print('test mat=',confusion_matrix(y_test,test_out))
print('f1=',f1_score(y_test,test_out))
sb.heatmap((confusion_matrix(y_test,test_out)),annot=True)
pt.xticks([0,1],['benign','melanoma'])
pt.yticks([0,1],['benign','melanoma'],rotation=0)
pt.show()
```
